```python
#화폐 단위 문자형을 입력받아 숫자형으로 반환 format 함수
#스크랩 중간에 디젤 값이 없는 주유소 발생,,
def thou_format(str):
    if s:
        return int(str.replace(",", "")) 
    else:
        return 0

def thou_format(s):
    if s:
        return int(s.replace(",", ""))
    else:
        return 0


#주유소 브랜드를 입력하면 gasbrand 데이터를 받아 id를 반환하는 함수
def station_id(brand):
    sql = "SELECT * FROM GAS_BRAND"
    cur.execute(sql)
    result = cur.fetchall()
    for i in result:
        if i[1] == brand:
            return i[0]
        elif brand == '알뜰(ex)':
            return 5
        elif brand == 'HD현대오일뱅크':
            return 2

#주소를 입력받아 구 이름을 반환하는 함수 split
def extrace_gu(address):
    return address.split(" ")[1]

#주소를 입력 받아 위도, 경도를 반환하는 함수 gmap
def get_latlng(address):
    tmp = gmaps.geocode(address, language = 'ko')
    lat = tmp[0].get('geometry')['location']['lat']
    lng = tmp[0].get('geometry')['location']['lng']
    
    return lat, lng
```

- 다음의 형식으로 저장되어야 함 (브랜드 이름, 칼럼 명 주의, id : GAS_STORE.id)

제출 5.
- 시각화 프로젝트 관련 코드 (ipynb), 결과 파일 (csv)

```python
#데이터를 가져올 때 브랜드 아이디 대신 브랜드명이 표시
sql = "SELECT s.id, b.name 'brand', s.name, s.city, s.gu, s.address, s.gasoline, s.diesel, s.self, s.car_wash, s.charging_station, s.car_maintenance, \
s.convenience_store, s.24_hours, s.lat, s.lng \
FROM GAS_STATION s, GAS_BRAND b \
WHERE b.id = s.brand ORDER BY s.id"

cur.execute(sql)
result = cur.fetchall()

for row in result:
    print(row)
```

```python
#csv 파일로 저장
num_fields = len(cur.description)
field_names = [i[0] for i in cur.description]

df = pd.DataFrame(result)
df.columns = field_names

df.to_csv("gas_station_output.csv", index = False, encoding = 'utf-8')
df.head()
```

문제 8.

구별로 주유소 브랜드 별 휘발유 평균가격을 조회하여 저렴한 순으로 출력하세요.

- 구 이름, 주유소 브랜드 이름, 휘발유 평균 가격

제출 8.
- 쿼리, 결과

```python
sql = "SELECT s.gu, b.name, AVG(s.gasoline) avg \
FROM GAS_BRAND b, GAS_STATION s \
WHERE b.id = s.brand GROUP BY s.gu, b.name ORDER BY s.gu, avg;"

cur.execute(sql)
result = cur.fetchall()

for row in result:
    print(row)
```
